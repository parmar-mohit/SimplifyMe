# Drawing Graphs

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel("Loss Data.xlsx")
df.head()
```

```python
plt.plot(df["Steps"], df["Training Loss"],label="Training Loss")
plt.plot(df["Steps"], df["Validation Loss"],label="Validation Loss")
plt.xlabel("Steps")
plt.ylabel("Loss")
plt.title("Steps vs Loss")
plt.legend()
plt.plot(figsize=(20,1))
plt.savefig("Loss Graph.png")
```
