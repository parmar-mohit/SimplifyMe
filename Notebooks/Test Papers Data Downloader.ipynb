# Test Papers Data Downloaders

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import requests
```

```python
def download_paper(paper_id):
    headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5)AppleWebKit 537.36 (KHTML, like Gecko) Chrome", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8"}
    download_url = 'http://ieeexplore.ieee.org/stampPDF/getPDF.jsp?tp=&isnumber=&arnumber={}'.format(paper_id)
    response = requests.get(download_url,headers=headers)
    if response.status_code == 200:
        with open("../Test Paper Data/"+paper_id+".pdf", 'wb') as f:
            f.write(response.content)
        print("Paper Id : {}, PDF downloaded successfully!".format(paper_id))
    else:
        print("Failed to download PDF. Status code:", response.status_code)
```

```python
site_url = "https://ieeexplore.ieee.org/search/searchresult.jsp?queryText=summarization&highlight=true&returnType=SEARCH&matchPubs=true&openAccess=true&returnFacets=ALL&pageNumber={}"

papers_to_download = 100
papers_downloaded = 0
page_number = 1

while papers_downloaded < papers_to_download:
    browser = webdriver.Chrome()
    browser.get(site_url.format(page_number))
    wait = WebDriverWait(browser, 10)
    wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "div.List-results-items")))
    html_content = browser.page_source
    browser.quit()

    soup = BeautifulSoup(html_content,"html.parser")
    papers = soup.find_all("div",{"class":"List-results-items"})

    for paper in papers:
        paper_ref = paper.find("a",{"aria-label":"PDF"})
        paper_id = paper_ref["href"].split("=")[-1]
        download_paper(paper_id)
        papers_downloaded += 1

        if papers_downloaded == 100:
            break
    page_number += 1

print("Papers Downloaded Successfully")
```
